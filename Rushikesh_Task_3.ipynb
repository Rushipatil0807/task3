```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv("/content/householdtask3.csv")
```

```python
display(data.head(10))
```

```python
#scatter plot
plt.scatter(data["year"],data["own"])

plt.title("scatter Plot")

plt.xlabel("year")
plt.ylabel("own")

plt.show()
```

```python
#line chart
plt.plot(data["year"]),
plt.plot(data["own"])

plt.title("line Plot")

plt.xlabel("year")
plt.ylabel("own")

plt.show()
```

```python
#Bar chart
plt.bar(data["year"],data["own"])

plt.title("Bar chart")

plt.xlabel("year")
plt.ylabel("own")

plt.show()
```

```python
#Histogram
plt.hist(data["income"])

plt.title("Histogram")

plt.show()
```
